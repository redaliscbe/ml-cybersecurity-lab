```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Create sample date values
dates = pd.date_range(start="2024-01-01", periods=120, freq="D")

# Create sample network traffic data
trend = np.linspace(1000, 1800, 120)
seasonal = 200 * np.sin(np.arange(120) * 2 * np.pi / 7)
noise = np.random.normal(0, 50, 120)

traffic = trend + seasonal + noise

# Store data in DataFrame
data = pd.DataFrame({
    "Date": dates,
    "Network_Traffic": traffic
})

data.set_index("Date", inplace=True)

# Decompose the time series
result = seasonal_decompose(
    data["Network_Traffic"],
    model="additive",
    period=7
)

# Plot clearly
fig = result.plot()
fig.set_size_inches(12, 8)

plt.suptitle("Time Series Decomposition of Network Traffic", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Read network traffic dataset
data = pd.read_csv("./dataset/network_traffic.csv")

# Convert Date column
data["Date"] = pd.to_datetime(data["Date"])

# Set Date as index
data.set_index("Date", inplace=True)

# Set daily frequency
data = data.asfreq("D")

# Fill missing values
data["Requests"] = data["Requests"].interpolate()

# Decompose network requests
result = seasonal_decompose(
    data["Requests"],
    model="additive",
    period=7
)

# Add components to dataframe
data["Trend"] = result.trend
data["Seasonal"] = result.seasonal
data["Residual"] = result.resid

# Detect abnormal spikes
threshold = data["Residual"].std() * 2
data["Anomaly"] = abs(data["Residual"]) > threshold

print(data)

# Plot traffic with anomalies
plt.figure(figsize=(10,4))
plt.plot(data.index, data["Requests"], label="Network Requests")
plt.scatter(
    data.index[data["Anomaly"]],
    data["Requests"][data["Anomaly"]],
    color="red",
    label="Possible Attack"
)

plt.title("DDoS Pattern Detection using Time Series Decomposition")
plt.xlabel("Date")
plt.ylabel("Requests")
plt.legend()
plt.show()
```
